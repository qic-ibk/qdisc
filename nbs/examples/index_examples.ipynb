# Readme

These examples reproduce the results presented in our paper *"Discovering Quantum phenomena with Interpretable Machine Learning"*, covering simulations across a range of quantum systems:

 - Experimental snapshots of [Rydberg atoms](qdisc_Rydberg.ipynb)
 - Classical shadows of the [cluster Ising model](qdisc_ClusterIsing.ipynb)
 - Hybrid data of a [fermionic system](qdisc_FKM.ipynb)

These are more advanced examples aimed to reproduce the results of our paper. For a more detailed walkthrough of the library, refer to the tutorials on [VAEs](../tutorials/J1J2_tuto1_VAE.ipynb) and [symbolic regression](../tutorials/J1J2_tuto2_SR.ipynb).

For exact reproducibility, the examples were run with the following package versions:

```
jax==0.4.38
jaxlib==0.4.38
netket==3.15.2
flax==0.10.4
optax==0.2.4
CUDA 12.4
```
