# Readme

The series of examples reproduce the results of the paper. It contains simulations for various types of data from different quantum systems:
 - experimental snapshots of [Rydberg atoms](qdisc_Rydberg.ipynb)
 - classical shadows of the [cluster Ising model](qdisc_ClusterIsing.ipynb)
 - hybrid data of a [fermionic system](qdisc_FKM.ipynb)

For a more detailed explanation of the library, please refer to the *tutorials*.

For exact reproducibility, the codes were run with: jax 0.4.38, netket 3.15.2, flax 0.10.4, optax 0.2.4, jaxlib 0.4.38, CUDA 12.4

The requirement.txt can be asked if needed.
